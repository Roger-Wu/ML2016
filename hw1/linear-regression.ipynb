```python
%pylab inline
import numpy as np
import pandas as pd
```

```python
DATA_FOLDER = "data/"
MODEL_FOLDER = "models/"
OUTPUT_FOLDER = "outputs/"
TRAIN_FILE_PATH = DATA_FOLDER + "train.csv"
TEST_FILE_PATH = DATA_FOLDER + "test_X.csv"
```

```python
# read train data
train_data = pd.read_csv(TRAIN_FILE_PATH, header=0, encoding="big5")
```

```python
# preprocess train data
train_data[train_data=="NR"] = 0
```

```python
train_data
```

```python
# concat 20 days data
trainDataByMonth = []  # 12 numpy.array, array's shape is (18, 480)
for month in range(12):
    trainDataArr = np.concatenate( [train_data[(360*month+18*day):(360*month+18*(day+1))][list(range(3, 27))].apply(pd.to_numeric).as_matrix() for day in range(20)], axis=1)
    trainDataByMonth.append(trainDataArr)
```

## Train

```python
lr = 0.1  # learning rate
```

```python
w = np.random.randn(18, 9)
b = np.random.randn()
```

```python
def singleTrain(trainDataByMonth, w, b):
    x = trainDataByMonth[0][:18, 0:9]
    predY = einsum('ij,ij', x, w) + b
    y = x * w
    print(y)
```

```python
w = np.array([[-0.05286123,  0.04785045,  0.17396246, -0.35730744,  0.22641066 , 0.32346691,
  -0.64073439 , 0.25740521 , 0.95267167]])  # np.random.randn(1, 9)
b = 1.31919712784  # np.random.randn()
```

```python
lr = 5 * 10**(-9)

print(w, b)
prevLoss = np.inf

for saveIdx in range(10):
    for epoch in range(1000):
        loss = 0
        deltaB = 0
        deltaW = np.zeros((1, 9))
        for month in range(12):
            for hour in range(470):
                x = trainDataByMonth[month][9:10, hour:hour+9]
                realY = trainDataByMonth[month][9, hour+9]

                predY = einsum('ij,ij', x, w) + b
                diff = realY - predY
                loss += diff * diff

                mult = 2 * diff * (-1)
                deltaB += mult
                deltaW += mult * x  # 2 * diff * (-1) * x

        loss /= (12*470)

        if loss > prevLoss:
            lr *= 0.75
        
        w = w - lr * deltaW
        b = b - lr * deltaB
        
        prevLoss = loss
    
    print(w, b, lr, loss)
    np.savez(MODEL_FOLDER + str(loss), w=w, b=b)

print('end')
```
