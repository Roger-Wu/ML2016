```python
import numpy as np
import pandas as pd
```

```python
DATA = "data/ml2016-pm2-5-prediction_public_leaderboard.csv"
```

```python
train_data = pd.read_csv(DATA, header=0)
```

```python
train_data.head()
```

```python
train_data['TeamName']
```

```python
train_data_extr = train_data[train_data['TeamName'].str.contains("_")]
```

```python
train_data_extr = train_data_extr.sort_values(['Score'])
```

```python
train_data_extr = train_data_extr.drop_duplicates('TeamName')
```

```python
train_data_extr = train_data_extr.reset_index(drop=True)
```
